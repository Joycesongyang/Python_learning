## Machine Learning of PurdueUniversity Exericise:  Plot function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = 1/(1 + np.exp(-5*(x-1)))
```

```python
plt.plot(x, y, linewidth = 6, color = 'black')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('my plot')
plt.show()
```
